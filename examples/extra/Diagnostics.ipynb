## Plotting Model Diagnostics

This notebook shows how to perform diagnostic checks using `hetGPy`. The diagnostic plot shows the observed vs predicted values, as well as the kriging variances.

```python
from hetgpy import hetGP
from hetgpy.example_data import mcycle
import numpy as np
%config InlineBackend.figure_formats = ['svg']
m = mcycle()
X, Z = m['times'], m['accel']
model = hetGP()
model.mle(
    X = X,
    Z = Z,
    covtype="Matern5_2",
    lower = np.array([1.0]),
    upper=np.array([10.0])
)
fig, ax = model.plot()
fig.tight_layout()
```
